```python
def plot_on_bar(a):
    a = pd.Categorical(a)
    length = len(a)
    plt.figure(figsize=(15,5))
    ax = sns.countplot(a,palette='Paired')
    for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/length)
        x = p.get_x() + p.get_width() / 2 - 0.05 
        y = p.get_y() + p.get_height()           
        ax.annotate(percentage, (x, y), size = 12)
    plt.show() 
```

```python
categorical_col = []
for column in df.columns:
    if df[column].dtypes != "float64" and df[column].dtypes != "int64":
        categorical_col.append(column)
le = LabelEncoder()
for col in categorical_col:
    df[col] = le.fit_transform(df[col])
pd.set_option("display.max_columns",None)


x = df.drop(["id","description","first_review","host_since","last_review","neighbourhood", "zipcode"],axis = 1)
x.dropna(axis=0, inplace=True)

y = x.log_price



x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=101)

lr = LinearRegression()

lr.fit(x_train,y_train)

y_predict_lr = lr.predict(x_test)

mae_lr = metrics.mean_absolute_error(y_test, y_predict_lr)
mse_lr = metrics.mean_squared_error(y_test, y_predict_lr)
rmse_lr = np.sqrt(metrics.mean_squared_error(y_test, y_predict_lr))
r2_lr = metrics.r2_score(y_test, y_predict_lr)

print('\nMean Absolute Error of Linear Regression: ', mae_lr)
print('\nMean Squarred Error of Linear Regression: ', mse_lr)
print('\nRoot Mean Squarred Error of Linear Regression: ', rmse_lr)
print('\nR2 Score of Linear Regression: ', r2_lr)
```

#### Plus le MAE est bas, meilleure est la performance du modÃ¨le. Dans ce cas, le MAE est d'environ 0.369.
#### Un MSE plus faible indique une meilleure performance du modÃ¨le. Dans ce cas, le MSE est d'environ 0.235.
#### Comme pour le MSE, un RMSE plus faible est prÃ©fÃ©rable. Dans ce cas, le RMSE est d'environ 0.484.
#### Une valeur plus Ã©levÃ©e de R2 indique une meilleure adÃ©quation du modÃ¨le aux donnÃ©es. Dans ce cas, le R2 Score est d'environ 0.547, ce qui indique que le modÃ¨le explique environ 54.7% de la variance dans la variable cible.
